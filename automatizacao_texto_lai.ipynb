```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('linhas_onibus_maua.csv')
df.head()
```

```python
df
```

```python
def solicitacaoSIC(row):
    textoSIC = 'Com base na LAI: a Lei de Acesso à Informação, LEI nº 12.527/2011, e seguindo a orientação da solicitação 2023/18 e do resposta a OGM 128/2023, solicito arquivo digital tabulado em formato aberto, que contenha a relação da quantidade de veículos (ônibus) programadas para a Linha @numerolinha - @nomelinha, do sistema de transporte municipal, pelos dias da semana e horários em que o sistema funciona'
    textoPersonalizado = textoSIC.replace('@numerolinha', str(row['linha'])).replace('@nomelinha', row['nome da linha'])
    return textoPersonalizado
```

```python
df['Solicitações SIC'] = df.apply(solicitacaoSIC, axis=1)
```
